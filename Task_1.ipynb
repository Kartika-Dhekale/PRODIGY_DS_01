This project analyzes global population data using visualization techniques.
A histogram is used to understand the distribution of population across countries,
while a bar chart highlights the top 10 most populated countries.
These visualizations provide insights into global population patterns.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)
df.head()
```

```python
population = df["2022"]

plt.figure()
plt.hist(population, bins=15)

plt.title("Population Distribution (2022)")
plt.xlabel("Population")
plt.ylabel("Frequency")

plt.show()
```

```python
top10 = df.sort_values(by="2022", ascending=False).head(10)

plt.figure()
plt.bar(top10["Country Name"], top10["2022"])

plt.title("Top 10 Countries by Population")
plt.xlabel("Country")
plt.ylabel("Population")

plt.xticks(rotation=45)
plt.show()
```
